## Adding Kaggle key to environment

```python
from google.colab import userdata
from os import environ

environ["KAGGLE_KEY"]      = userdata.get('KAGGLE_KEY')
environ["KAGGLE_USERNAME"] = userdata.get('KAGGLE_USERNAME')

print("key added to environ") if (environ["KAGGLE_KEY"] and
                                  environ["KAGGLE_USERNAME"]) else print("Could not find the env")
```

# Importing MitraLLM

> [ref](https://colab.research.google.com/drive/1OdDSY5hr67Frqj0s8_TzUVFe8e7CaSB3)

>NOTE: Make sure you have joined Discord, and complited onboarding. You will get your `TOKEN` and `ACCESSAPI` after successful onboarding.

>Add the `TOKEN` and `ACCESSAPI` to colab secrets

| Key       | Value             |
| ----------|-------------------|
| TOKEN     | *copied token*    |
| ACCESSAPI | *copied accessapi*|

```python
!pip install git+https://gitlab.com/gauthammsam/mitrallm.git --quiet
```

```python
from mitrallm import MitraLLM
from google.colab import userdata

llm = MitraLLM(
    token     = userdata.get("MITRA_TOKEN"),
    accessapi = userdata.get("MITRA_ENDPOINT")
) ## llm is a langchain model
print(llm)
```

```python
response = llm.invoke("How old is the earth? Tell me in one sentence.")
response
```
